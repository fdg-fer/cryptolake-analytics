```python
%sql
CREATE OR REPLACE TABLE uc_sa_br_dev.gold_finance_investments_market_analysis.dim_datetime
USING DELTA
AS
SELECT
  datetime_value AS datetime_id,
  TO_DATE(datetime_value) AS date_id,
  datetime_value AS datetime,

  DATE_FORMAT(datetime_value, 'dd/MM HH:00') AS datetime_label,

  YEAR(datetime_value) AS year,
  QUARTER(datetime_value) AS quarter,
  MONTH(datetime_value) AS month,
  DAY(datetime_value) AS day,
  HOUR(datetime_value) AS hour,
  DATE_FORMAT(datetime_value, 'yyyy-MM') AS year_month,

  DAYOFWEEK(datetime_value) AS day_of_week,

  CASE DAYOFWEEK(datetime_value)
    WHEN 1 THEN 'Domingo'
    WHEN 2 THEN 'Segunda'
    WHEN 3 THEN 'Terça'
    WHEN 4 THEN 'Quarta'
    WHEN 5 THEN 'Quinta'
    WHEN 6 THEN 'Sexta'
    WHEN 7 THEN 'Sábado'
  END AS day_of_week_name,

  CASE 
    WHEN DAYOFWEEK(datetime_value) IN (1, 7) THEN TRUE
    ELSE FALSE
  END AS is_weekend,

  CASE
    WHEN HOUR(datetime_value) BETWEEN 0 AND 5 THEN 'Madrugada'
    WHEN HOUR(datetime_value) BETWEEN 6 AND 11 THEN 'Manhã'
    WHEN HOUR(datetime_value) BETWEEN 12 AND 17 THEN 'Tarde'
    ELSE 'Noite'
  END AS hour_period

FROM (
  SELECT EXPLODE(
    SEQUENCE(
      TIMESTAMP('2024-01-01 00:00:00'),
      TIMESTAMP('2027-12-31 23:00:00'),
      INTERVAL 1 HOUR
    )
  ) AS datetime_value
);
```
